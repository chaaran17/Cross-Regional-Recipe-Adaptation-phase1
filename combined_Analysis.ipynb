```python
# DEGREE CENTRALITY 
'''calculating the degree centrality - Degree centrality gives us the information about the degree of the nodes 
A high degree centrality score simply means that a node has more connections'''
```

```python
degree_centrality = nx.degree_centrality(G)
```

```python
# Print degree centrality for each node
for node, centrality in degree_centrality.items():
    print(f"Node {node}: Degree Centrality {centrality}")
```

```python
# Top 20 nodes Acc to degree centrality
top_20_nodes = sorted(degree_centrality, key=degree_centrality.get, reverse=True)[:20] 
```

```python
top_20_nodes
```

```python
for node in top_20_nodes:
    print(f"Node {node}: Degree Centrality {degree_centrality[node]}")
```

```python
# Create lists for plotting
nodes = []
centrality_values = []
```

```python
# Extract nodes and their corresponding centrality values
for node in top_20_nodes:
    nodes.append(node)
    centrality_values.append(degree_centrality[node])
```

```python
# Plot the degree centrality values
plt.figure(figsize=(10, 6))
plt.bar(nodes, centrality_values, color='skyblue')
plt.xlabel('Nodes')
plt.ylabel('Degree Centrality')
plt.title('Degree Centrality of Top 20 Nodes')
plt.xticks(rotation=45)
plt.show()
```

```python
# COMMUNITY DETECTION
import pandas as pd
import networkx as nx
import community as community_louvain
import matplotlib.pyplot as plt
```

```python
# Create a NetworkX graph
G = nx.from_pandas_edgelist(df, source='Source', target='Target', edge_attr=True)
```

```python
# Perform community detection using the Girvan-Newman algorithm
comp = nx.algorithms.community.centrality.girvan_newman(G)
```

```python
# Get the communities
communities = tuple(sorted(c) for c in next(comp))
```

```python
for i, community in enumerate(communities):
    print(f"Community {i+1}:")
    print(community)
```

```python
# Draw the graph with node colors based on communities
pos = nx.spring_layout(G)  # you can use a different layout algorithm
plt.figure(figsize=(10, 8))
for i, community in enumerate(communities):
    nx.draw_networkx_nodes(G, pos, nodelist=community, node_size=50, node_color=plt.cm.tab20b(i))
nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.show()
```

```python
# Compute edge betweenness centrality
edge_betweenness = nx.edge_betweenness_centrality(G)
```

```python
# Sort edges based on betweenness centrality
sorted_edges = sorted(edge_betweenness, key=edge_betweenness.get, reverse=True)
```

```python
# Print the top 10 key connectors
print("Top 10 Key Connectors:")
for edge in sorted_edges[:10]:
    print(edge)
```

```python
# Draw the graph with key connectors highlighted
pos = nx.spring_layout(G)  # you can use a different layout algorithm
plt.figure(figsize=(10, 8))
nx.draw_networkx_nodes(G, pos, node_size=50)
nx.draw_networkx_edges(G, pos, edgelist=sorted_edges[:10], edge_color='r', width=2)
nx.draw_networkx_edges(G, pos, edgelist=sorted_edges[10:], edge_color='b', alpha=0.5)
plt.show()





```
